```python
print("\n--- Describe Data ---")
df.describe()
```

```python
print("\n---All Value Counts ---")

df.value_counts()
```

```python

print("\n---Specific Value Counts ---")
print("Sex:")
print(df['Sex'].value_counts())
print("\nEmbarked:")
print(df['Embarked'].value_counts())
```

```python
print("\n--- Missing Values ---")
df.isnull().sum()
```

```python
# --- Correlation Heatmap ---
df_numeric = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10,8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# --- Pairplot ---
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# --- Histograms ---
df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

df['Fare'].hist(bins=30, edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# --- Boxplots ---
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age vs Passenger Class')
plt.show()

sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()

```

```python
# --- Scatterplot ---
plt.scatter(df['Age'], df['Fare'], c=df['Survived'], cmap='coolwarm', alpha=0.6)
plt.title('Age vs Fare (colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
print("""
Observations:
- Majority of passengers did not survive.
- Females had a higher survival rate compared to males.
- Passengers in 1st class had a higher survival chance.
- Higher fares were associated with higher chances of survival.
- There are missing values in Age, Cabin, and Embarked columns.
- Fare distribution is right-skewed with some very high values (outliers).
- Age distribution is fairly normal but with missing data.
- Strong negative correlation between Pclass and Fare.

Summary:
Overall, survival is influenced by sex, passenger class, and fare. Females, higher class passengers, and those who paid higher fares had better chances of survival.
""")
```
